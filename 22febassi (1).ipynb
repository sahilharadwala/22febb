Q6. Which of the following is mutable in nature Series, DataFrame, Panel?
In Pandas, both Series and DataFrame are mutable, meaning that the values they contain can be changed: 
Series
The value of an element in a Series can be changed after it has been initialized. However, the length of a Series cannot be changed. 
DataFrame
The values contained in a DataFrame can be changed, and columns can be inserted or deleted. A DataFrame can be thought of as a dictionary of Series structures
